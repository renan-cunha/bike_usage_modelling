Todos os experimentos aqui rodados será executado 1000 vezes pela característica estocástica do problema. Para plotar os gráficos, será usada a média dessas 1000 execuções. Além disso, sempre será considerado o período de 24 horas para a simulação.

```python
from simulation import Simulation
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def run_simulation(time, docks, bikes):
        if bikes > docks:
            print("The number of bikes cannot be greater than the number of docks.")
            return
        simulation = Simulation(bikes, docks-bikes, time)
        results = simulation.start_simulation()
        return results
```

## Cenário 1
    - 70 docks fixas 
    - Bikes variando [35, 70]

```python
time = 24
docks = 70
n_bikes = []
taking = []
leaving = []

for bikes in range(docks//2, docks + 1):
    for i in range(1000):
        results = run_simulation(time, docks, bikes)
        n_bikes.append(bikes)
        taking.append(results[1])
        leaving.append(results[2])
        
df = pd.DataFrame()
df['n_bikes'] = n_bikes
df['w_taking'] = taking
df['w_leaving'] = leaving
```

```python
sns.lineplot(data=df, x='n_bikes', y='w_taking', label='alugar bike')
sns.lineplot(data=df, x='n_bikes', y='w_leaving', label='devolver bike')
plt.xlabel('Bicicletas')
plt.ylabel('Desistência');
```

## Cenário 2
    - 70 docks variando [10, 70] 
    - Bikes sempre sendo a metade do número de docks [5, 35]

```python
time = 24
min_docks = 10
max_docks = 70
n_docks = []
taking = []
leaving = []

for dock in range(min_docks, max_docks + 1):
    bikes = dock//2
    for i in range(1000):
        results = run_simulation(time, dock, bikes)
        n_docks.append(dock)
        taking.append(results[1])
        leaving.append(results[2])
        
df2 = pd.DataFrame()
df2['n_docks'] = n_docks
df2['w_taking'] = taking
df2['w_leaving'] = leaving
```

```python
sns.lineplot(data=df2, x='n_docks', y='w_taking', label='alugar bike')
sns.lineplot(data=df2, x='n_docks', y='w_leaving', label='devolver bike')
plt.xlabel('Docks')
plt.ylabel('Desistência');
```
